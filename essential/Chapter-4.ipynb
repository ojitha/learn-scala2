## Traits

- no constructor
- can have abstract methods without implementations

Traits are templates for creang classes, in the same way that classes are templates for creang objects.


````scala
import java.util.Date

trait Visitor {
    def id: String
    def createdAt : Date
    def age: Long = new Date().getTime - createdAt.getTime 
}

case class User (val id: String
               , val email: String
               , val createdAt: Date = new Date()) extends Visitor
````

> `def` is a more general version of the `val`

````scala
def isOlder(v1: Visitor, v2: Visitor): Boolean 
    = v1.createdAt.before(v2.createdAt)

val visitor1 = User ("1", "v1@test.com", new Date())
val visitor2 = User ("2", "v2@test.com", new Date())

isOlder(visitor1, visitor2)
````

## This or That and Nothing Else: Sealed Traits

When we mark a trait as sealed, we must define all of its subtypes in the same file.

> Once the trait is sealed, the compiler knows the complete set of subtypes and will warn us if a pattern-matching expression is missing. We will not get a similar warning from an unsealed trait.

We can still extend the subtypes of a sealed trait outside of the file where they are defined. If we want to prevent this possibility, we should declare them as sealed (if we want to allow extensions within the file) or final if we want to disallow all extensions.

## Modelling Data with Traits

- sum type pattern (is a) : Or
- product type pattern (has a) : And

### Algebraic Data Types
An algebraic data type is any data that uses the above two patterns. In the functional programming literature, data using the “has-a and” pattern is known as a product type, and the “is-a or” pattern is a sum type.

“is-a and” pattern means that A is a B and C

````scala
trait B
trait C
trait A extends B with C
````

“has-a or” patterns mean that `A` has `B` or `C`.

1. A has a `d` of Type `D`, where `D` is a `B` or `C`
   ```scala
   trait A {
       def d: D
   }
    
   sealed trait D
   final case class B() extends D
   final case class C() extends D
   ```

2. Alternatively, `A` is `D` or `E`, and `D` has a `B` and `E` has `C`
   ```scala
   sealed trait A
   final case class D(b: B) extends A
   final case class E(c: C) extends A
   ```

## Working With Data

> The sum and product type patterns tell us how to combine data to make bigger data. 

**Structural recursion** says that if we have an A as defined before, we must break it into its constituent `B` and `C`, which we then
combine in some way to get closer to our desired answer. 

> Structural recursion is essentially the process of breaking down data into smaller pieces.

Two patterns for decomposing them using structural recursion:

1. Structural Recursion using Polymorphism
2. Structural Recursion using Pattern Matching

###  Structural Recursion using Polymorphism
example:

````scala
sealed trait A {
   def foo: String 
}

final case class B() extends A {
    def foo: String = "It's B!"
}

final case class C() extends A {
    def foo: String = "It's C"
}
````

````scala
val anA: A = B()
anA.foo
````

````scala
val anA: A = C()
anA.foo
````

We can define implementation in trait and override the method in B and C:

````scala
sealed trait A {
   def foo: String = "It's A!"
}

final case class B() extends A {
    override def foo: String = "It's B!"
}

final case class C() extends A {
    override def foo: String = "It's C"
}
````

````scala
val anA: A = B()
anA.foo
````

### Structural Recursion using Pattern Matching

````scala
trait F // target

def f(a: A): F = a match {
    case B() => ???
    case C() => ???
}
````

For example

````scala
sealed trait Food
case object Antelope extends Food
case object TigerFood extends Food
case object Licorice extends Food
final case class CatFood(food: String) extends Food
````

Using pattern matching:

````scala
sealed trait Feline {
    def dinner: Food = this match {
        case Lion() => Antelope
        case Tiger() => TigerFood
        case Panther() => Licorice
        case Cat(favouriteFood) => CatFood(favouriteFood)
    }
}
final case class Lion() extends Feline
final case class Tiger() extends Feline
final case class Panther() extends Feline
final case class Cat(favouriteFood: String) extends Feline
````

Using pattern matching using external object

````scala
object Diner {
    def dinner(feline: Feline): Food = feline match {
        case Lion() => Antelope
        case Tiger() => TigerFood
        case Panther() => Licorice
        case Cat(favouriteFood) => CatFood(favouriteFood)
    }
}
````
