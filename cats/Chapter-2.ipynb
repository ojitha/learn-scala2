```scala
trait Monoid[A] {
    def combine(x: A, y: A): A
    def empty: A
}
```

```scala
def associativeLaw[A](x: A, y: A, z: A)(implicit m: Monoid[A]): Boolean = {
    m.combine(x, m.combine(y, z)) == m.combine(m.combine(x, y),z)
}

def identityLaw[A](x: A)(implicit m: Monoid[A]): Boolean = {
    (m.combine(x, m.empty) == x) && (m.combine(m.empty, x) == x)
}
```

A semigroup is just the combined part of a monoid without the empty part. Therefor Monoid can be define as:

```scala
trait Semigroup[A] {
    def combine (x: A, y: A): A
}

trait Monoid[A] extends Semigroup[A] {
    def empty: A
}

object Monoid {
    def apply[A](implicit monoid: Monoid[A]) = monoid
}
```

Consider `Boolean`, how many monoids for this type?

> For each monoid, define the combine and empty operations and convince yourself that
the monoid laws hold.



```scala
implicit val booleanAndMonoid: Monoid[Boolean] = new Monoid[Boolean] {
    def combine(a: Boolean, b: Boolean) = a && b
    def empty = true
}
```

```scala
implicit val booleanOrMonoid: Monoid[Boolean] = new Monoid[Boolean] {
    def combine(a: Boolean, b: Boolean) = a || b
    def empty = true
}
```

```scala
implicit val booleanEitherMonoid: Monoid[Boolean] = new Monoid[Boolean] {
    def combine(a: Boolean, b: Boolean) = (a && !b) || (!a && b)
    def empty = true
}
```

```scala
implicit val booleanXnorMonoid: Monoid[Boolean] = new Monoid[Boolean] {
    def combine(a: Boolean, b: Boolean) = (a || !b) && (!a || b)
    def empty = true
}
```

Cats provides syntax for the combine method in the form of the |+| operator.
Because combine technically comes from Semigroup

```scala
import cats.instances.string._
import cats.syntax.semigroup._

val stringResult = "Hi " |+| "there" |+| Monoid[String].empty
```

```scala
import cats.instances.map._

val map1 = Map("a" -> 1, "b" -> 2)
val map2 = Map("c" -> 3, "d" -> 4)

map1 |+| map2
```

```scala
import cats.instances.tuple._

val t1 = ("Hello", 1)
val t2 = (" " , 2)
val t3 = ("world", 3)

t1 |+| t2 |+| t3
```

We can also write generic code that works with any type for which we have
an instance of Monoid:

```scala
def addAll[A](values: List[A])(implicit monoid: Monoid[A]): A
    = values.foldRight(monoid.empty)(_ |+| _)

addAll(List(1,2,3))
addAll(List(None, Some(1), Some(2)))
```
